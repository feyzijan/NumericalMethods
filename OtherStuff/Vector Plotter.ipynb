```python
import numpy as np
import matplotlib.pyplot as pp
from mpl_toolkits import mplot3d
```

```python
# set ranges
x = np.arange(-5,6,1)
y = np.arange(-5,6,1)

Xg, Yg = np.meshgrid(x,y)

# 2 because vector has two components
F = np.ndarray((len(x),len(x),2))


# set i and j components
F[:,:,0] = Yg/(Xg**2 + Yg**2)
F[:,:,1] = -Xg/(Xg**2 + Yg**2)


pp.quiver(Xg,Yg,F[:,:,0],F[:,:,1], color="b")
```

```python
x = np.arange(-5,6,1)
y = np.arange(-5,6,1)

Xg, Yg = np.meshgrid(x,y)

# 2 because vector has two components
F = np.ndarray((len(x),len(x),2))


# set i and j components
F[:,:,0] = 1 + 0.1*Xg
F[:,:,1] = -0.1*Yg


pp.quiver(Xg,Yg,F[:,:,0],F[:,:,1], color="b")
```
