```python
import math as mt
Dx = 0.1
x = np.arange(np.pi*-2, np.pi*2 + Dx,Dx)
y = np.arange(np.pi*-1, np.pi*2 + Dx,Dx)

#create grids
xg, yg = np.meshgrid(x,y)


# evaluate functions
f = np.sin(xg) * np.cos(yg)
g = np.cos(xg) * np.sin(yg)
print(f)
print(g)

```

```python
# Question 2
s = f + g
p = f * g
print(s)
print(p)
```

```python
#Task C: Surface plots

# C1
# plot surface of s
# set 3D view 

ax = pp.axes(projection="3d")

ax.plot_surface(xg,yg,s)

# plot contour of s
# make a new window plot
pp.show()
# plot contours
pp.contour(xg,yg,s)

# lets do the same for p
# make a new window plot
pp.show()
# set 3d view
ax = pp.axes(projection="3d")
# now plot the surface
ax.plot_surface(xg,yg,p)

# plot contour of p
# make a new plot
pp.show()
# plot contour
pp.contour(xg,yg,p)
```

```python
# C2

# Define t range
Dt = 0.05 # set space
t = np.arange(0,10+Dt,Dt)

# set meshgrids
Xg, Yg, Tg = np.meshgrid(x,y,t)
# compute r
r =  np.sin(Xg) * np.cos(Yg) * np.exp(-0.5*Tg)
# C3
# plot r at t= 0
pp.show()
ax = pp.axes(projection="3d")
ax.plot_surface(Xg[:,:,0],Yg[:,:,0],r[:,:,0])

#now plot r at t = 5 ( halfway the range)

th = int(len(t)/2) # the index of the middle of t
pp.show() # you must write this to create a new plot
ax = pp.axes(projection="3d")
ax.plot_surface(Xg[:,:,th],Yg[:,:,th],r[:,:,th])

```

```python
#Task E: Bonus (a bit challenging) Slicing with conditions

dx = 0.1
x = np.arange(-5,5+dx,dx)
y = np.sin(x)

# length of y
N = len(y)

# compute ym
ym = np.zeros(N)
ym[y<0]= -y[y<0] # set absolute value conditon so that ym = |y|
ym[y>=0]= y[y>=0]

pp.plot(x,y,'r')
pp.plot(x,ym,'b+')


# E2
ymsat = np.zeros(N)
ymsat[ym<=0.5]=ym[ym<=0.5]
ymsat[ym>0.5]=0.5
ymsat[x<=0]=0
pp.plot(x,ymsat,'mx')
```
