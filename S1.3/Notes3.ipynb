## Counted Loops 

We have seen how to traverse lists by using the for loop operator. Repeated here, an example of printing elements of a list:

```python
Royals = ['Elisabeth', 'Charles', 'Camilla', 'William', 'Kate', 'Henry', 'Meghan', 'George', 'Louis', 'Charlotte']
for members in Royals: 
    print(members)
```

However, the for loop can be used, more in general, every time we wish to repeat a set of statements.

If a player wishes to throw a dice N times, we could rewrite the same statement of throwing a dice N times, or use a for loop to repeat the same action N times.
Every time the the same action is repeated. 
In order to ascertain how many time the action has been repeated, we need to count every occurrence of it. Every time we throw a dice we need to add one occurrence to our counting. In order to count, we would need a variable devoted to counting the number of repetition. This variable is usually known as the counter.


In addition of a counter, when repeating actions we need to know the intial and final counting, i.e. where we start counting from and up to which value.
It is therefore necessary to set beforehand a range from the beginning to the end of the counting.

```python
# count from
start = 10
# count up to 
end = 100
R = range(start,end+1)
```

When using the for loop, we set a variable devoted to counting within the specified range:

```python
for count in R:
    dice = int(random.random()*6 + 1)
    print(dice)
```

The variable count in the above example will act as the counter and will keep counting within the specified range. Every time that the actions within the loop are repeated, count is automatically incremented, as dictated by the range.

The counting does not need to be necessarily in unitary steps or forward. We could also count backwardly and with any integer step we wish or need.

```python
# count from end to start backwardly in unitary steps
R = range(end, start-1,-1)
# count upward in steps of 5
R = range(start,end+1,5)
print (R)
```

## Conditional statements

It is very common place in coding that the execution of some actions and/or statements will depend upon the verification of specific conditions, prescribed by the code developer.
For example, if two players P1 and P2 throw a dice each, the winner will depend of the player who has obtained the highest score. So, if we want to declare the winner, we need to verify which player has obtained the large score, and consequently declare either P1 or P2 as the winner.


Assessing who has obtained the highest score requires two steps:
1) Establishing a condition that would select either A or B as the winner, and
2) Verifying the condition

So, for example, we could enquire if the dice of P1 has a higher score than the dice of P2. The condition would then be "Is score of P1 higher than score of P2?".
Upon verification of this condition, we can either declare P1 as the winner if the condition is True, or P2 as the winner if such condition is False.

In Python this is implemented with the If operator:

```python
if dice1 > dice2:
    print('The winner is P1')
else:
    print('The winner is P2')
```

Of course, before posing the query, we would need to assign some values to the dices (as if they were thrown independently). 

```python
# load the random library
import random
# invoke the random function and assign the value generated to variable dice1
dice1 = int(random.random()*6 + 1)
print(dice1)
#  we do the same for dice2
dice2 = int(random.random()*6 + 1)
print(dice2)
```

Once the two dice have been thrown, i.e. dice1 and dice2 have a value assigned, we can enquire, verify the condition and declare the winner:


```python
if dice1 > dice2:
    print('The winner is P1')
else:
    print('The winner is P2')
```

The expression after the if is the condition. If the condition is True, the first set of intendent statements will run, otherwise the second set of indented statements will be executed.
The two possible set of statements are also known as branches.
There is no limit on the number of statements that can be executed within each branch, but there must be at least one.
Nevertheless, sometimes it may be convenient to have a branch with no statements, as a place keeper. In that case we can use the pass statement, that does nothing.

```python
mark = 79
if mark > 40:
    pass
else:
    print('I need to study more')
```

To condition is of Boolean type, i.e. it has only two possibilities. Any Boolean conditions is either True or False, unambiguously, and there cannot be a third possible branch.


### Boolean statements 

Boolean conditions can be stated with Boolean logical operators:


The use of logical operators 'and', 'or' and 'not' will allow for complex compounded conditions.
Example: assess if a number x is included within a range [a,b]:

```python
x = 11
a = 9
b = 20
if x >= a and x <= b:
    print('In range')
else:
    print('Out of range')
```

Incidentally, for this form of condition Python provides a more coincise format. However, do not abuse much of it, as it is easy to run into errors when the condition is more complex.

```python
x = 9
a = 9
b = 20
if a <= x < b:
    print('In range')
else:
    print('Out of range')
```

### Nested conditionals 

In assessing the winner between the two dice players, we have not considered the case when the two throws are identical: in such case there is no winner and the match is a draw.


Since, each Boolean condition can have only two branches, in order to consider the third option of the draw, we need to reformulate the enquire into Boolean statements.
We could firstly enquire if there is a winner of whether the match is a draw. Once establised that there is a winner, we could enquire, similarly to what done before, which of the two players has obtained the highest score and declare the winner.


To implement such logical conditions we need to nest an if statement within another if statement.

```python
dice1 = 1
dice2 = 3
if dice1 == dice2:
    print('The match is a draw')
else:
    print('Surely there is a winner')
    if dice1 > dice2:
        print('The winner is P1')
    else:
        print('The winner is P2')
```

### Common statements in branches

Let us create a list of 12 elements, containing the number of days for each month of the current year calendar.
The set of statements to create the list is straightfoward, except for the month of February, for which we would need to evaluate if it is a leap year or not.


After assessing if February contains 28 or 29 days, we would continue assigning all the other days from March to December. 


Since the days in months March - December is irrespective of what is happening in February, it would be redundant to repeat them twice in both the two possible branches of the condition. By doing so we are writing twice as much the number of statements needed.
It would be more sensible to extract out of the conditional construct any statement in common within the two branches.
