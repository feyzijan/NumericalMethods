# Task A

```python
# Trace of a matrix
def Trace(A):
    N = len(A) # determine the length of the matrix
    RN = range(0,N)
    t = 0 # initialise the trace
    for i in RN:
        # add the elements on the diagonal
        t += A[i][i]
    return t

# main
A = [[1,2,3,0],[4,5,6,0],[7,8,9,10], [4,5,6,2]]
t = Trace(A)
for row in A:
    print(row)
print(t)
```

# Task B

```python
# Matrix-Matrix multiplication
def MatMat(A,B):
    M = len(A) # rows of A
    NA = len(A[0]) # columns of A
    NB = len(B) # rows of B
    P = len(B[0]) # columns of B
    # define ranges
    RM = range(0,M)
    RN = range(0,NA)
    RP = range(0,P)
    # check that dimensions are compatible
    if NA == NB:
        # compute the product C = A * B
        C = []
        for i in RM:
            # compute row i of the matrix
            row = []
            for j in RP:
                # compute column j, i.e. element C[i][j]
                Sum = 0
                for k in RN:
                    Sum += A[i][k] * B[k][j]
                row += [Sum] # append this column
            C += [row] # append this row
    else:
        # dimensions not compatible
        C = 0
    return C

# main
# A 3x6
A = [[1,2,3,4,5,6],[6,0,3,4,2,1],[2,1,9,4,6,7]]
# B 6x4
B = [[1,2,3,4],[6,0,3,4],[2,1,9,4],[5,3,6,4],[1,1,0,5],[8,7,6,4]]
# C is 3x4
C = MatMat(A,B)
for row in C:
    print(row)
C = MatMat(B,A)
print(C)
```

# Task C

```python
# One possibilty could be sorting the two strings and then comparing if they are exactly the same.
# However, strings are immutable and we cannot change them.
# Hence, I am not going to follow this approach

# I'll follow this logic instead, though there may be many other valid logics too:
# Firstly I check if the two strings have same length, then
# I'll create a list of boolean values, setting all of them to False initially.
# This list is long as the characters in the strings.
# Then I'll traverse s1 and check if each character of it is in s2 too, 
# making sure that the character in s2 has not been counted already
# I'll verify this latest conditon by examining the Boolean list.
def IsAnagram(s1,s2):
    # check lengths
    ls1 = len(s1)
    ls2 = len(s2)
    if ls1 == ls2:
        # define the boolean list
        Bl = []
        for letter in s1:
            Bl += [False]
        # traverse s1
        i = 0
        Anagram = True
        while Anagram and i < ls1:
            # check if this character is in s2 too
            j = 0
            found = False
            while not found and j < ls2:
                # if found, make sure it wasn't considered already before
                if s1[i] == s2[j] and Bl[j] == False:
                    found = True # the character has been found in s2
                    Bl[j] = True # now mark this character in s2 as having been considered
                else:
                    # check next character in s2
                    j += 1
            if not found:
                # a character in s1 has not been found in s2, hence they do not anagram
                Anagram = False
            else:
                # examine next character in s1
                i += 1
    else:
        Anagram = False
        
    return Anagram
        
# main
s1 = 'bored'
s2 = 'robed'
c = IsAnagram(s1,s2)
print(c)
```

```python
import matplotlib.pyplot as pl
# read the data
f = open('Temperatures.txt','r')
t = f.readlines()
f.close()
# extract data
RD = range(0,10)
RH = range(0,24)

# compute daily averages and, at same time, find max and min
Av = [] # list of daily averages
days = [] # list of days
MaxT = int(t[0].rstrip()) # set max and min to first value
MinT = MaxT
# traverse all data, sequentially, and subdivide them in daily batches
c = 0
for day in RD:
    # examine one day
    days += [day]
    Sum = 0 # initialise daily average
    for hour in RH:
        # extract single temperature
        T = int(t[c].rstrip())
        # add this temperature for the average
        Sum += T
        # check if max or min
        if T > MaxT:
            MaxT = T
        if T < MinT:
            MinT = T
        # prepare to go into next data  
        c += 1
    # compute average for this day  
    Av += [Sum/24]

# print / plot
print(MinT,MaxT)
pl.bar(days,Av)
    
```

```python
# load the map
f = open('Map4.txt','r')
t = f.readlines()
f.close()
M = int(t[0].rstrip())
N = int(t[1].rstrip())
RM = range(0,M)
RN = range(0,N)
# for the matrix
Map = []
c = 2
for i in RM:
    row = []
    for j in RN:
        row += [int(t[c].rstrip())]
        c += 1
    Map += [row]

# Set the increments/decrements from current position (r,c) to access neighbouring cells
# NW N NE W E SW S SE
rnblist = [-1, -1, -1, 0, 0, 1, 1, 1]
cnblist = [-1, 0, 1, -1, 1, -1, 0, 1]

# set the initial points
# the first move will be inevitable into (1,1)
r = 1 # row
c = 1 # column
# therefore, the previous step was in (0,0)
rprev = 0
cprev = 0

# start moving along the path
# assume that there is a step forward
MoveForward = True
# keep mooving while there is a step forward
while MoveForward:
    # search in nearby position of the current position (r,c)
    # assume that there are no other stones to go further
    MoveForward = False
    i = 0
    # look at the 8 neighbouring cells, unless next stone is found
    while (not MoveForward ) and i < 8:
        # examine the neighbours to detect any move forward
        rnb = r + rnblist[i]
        cnb = c + cnblist[i]
        i = i + 1
        # if there is a stone and it is not the previous one
        # (the one we were coming from)
        if Map[rnb][cnb] == 0 and  not (rnb == rprev and cnb == cprev):
            # (rnb,cnb) is the next stone where to move
            # set the current position as the previous stone and ...
            rprev = r
            cprev = c
            # ... move forward
            r = rnb
            c = cnb
            # signal that there is a step forward
            MoveForward  = True


for row in Map:
    print(row)
# dislay the last stone visited
print(r,c)
```

# Task F

```python
def Hanoi(N, source, dest, spare):
    # This function moves N disks from peg source to peg dest
    # source: where to move from
    # dest: where to move to
    # spare: spare peg
    if N == 1:
        # move disk from source to dest
        # this is an obvious move
        print('move disk '+str(N)+' from '+source+' to '+dest)
    else:
        # move all top stack, but larger one, from source to spare
        Hanoi (N-1, source, spare, dest)
        # since the top stack has gone, we can move the larger disc directly from source to dest
        print('move disk '+str(N)+' from '+source+' to '+dest)
        # move all top stack from spare to dest, on top of the larger disc which has just been moved on to dest
        Hanoi (N-1, spare, dest, source)

# move 4 disks from A to C
Hanoi(3,'A','C','B')
```
