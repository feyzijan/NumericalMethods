```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
# Task A
def MyGauss(A,b):
    
    # number of equations
    n = len(b)
    
    # eliminate the unknowns, from first to (n-1)th unknown, to form an upper triangular matrix
    for i in range(0,n-1):
        # eliminate the i-th unknown from the (i+1)th row downwards
        # i.e. set the zeros in column i.
        for j in range(i+1,n):
            # eliminate on row j

            # A(i,i) is the pivot coefficient
            p = A[j,i] / A[i,i]
        
            # compute the new elements of row j in matrix A
            # use slicing
            #A[j,:] = A[j,:] - p * A[i,:]
            # or, alternatively, loop for every cell of row j
            for k in range(i,n):
                A[j,k] = A[j,k] - p * A[i,k]


            # compute the new element of row j in vector b
            b[j] = b[j] - p * b[i]
    
    # verify A has been reduced to a triangular matrix
    #print(A)
    
    # evauate, by back substitution the solution
    # start from the last unknown and go upward till the first unknown
    x = np.zeros(n)
    for i in range(n-1,-1,-1):
        # contribution from b (right hand side of the equation)
        x[i] = b[i] / A[i,i]
        # contribution from the other (already evaluated) unknowns
        # (within the left hand side of the equation)
        for k in range(i+1,n):
            x[i] = x[i] - A[i,k] * x[k] / A[i,i]

    return x
```

```python
# Task A2

A = np.array([[8, -2, 1, 3], [1, -5, 2, 1],[-1, 2, 7, 2],[2 ,-1, 3, 8]],dtype=float)
b = np.array([9, -7, -1, 5],dtype=float)

x = MyGauss(A,b)

print(x)
```

```python
# Task B1
A = np.array([[4,3],[8,6]])
b = np.array([2,1])

xp = np.arange(-5,5,0.1)
y1 = b[0]/A[0,1]-A[0,0]/A[0,1]*xp
y2 = b[1]/A[1,1]-A[1,0]/A[1,1]*xp

pl.plot(xp,y1,'r')
pl.plot(xp,y2,'b')

x = MyGauss(A,b)
print(x)

det = np.linalg.det(A)
print(det)
```

```python
# Task C
# the set of equations for the circuit given is not unique:
# it depends of the KVL loops and KCL nodes that you choose to include in the set
R1 = 50.0
R2 = 100.0
R3 = 50.0
R4 = 100.0
R5 = 50.0
V = 1.0

# my KVL, KCL set:
# R2i2 = R1i1 + R3i3
# i1 + i2 = i4 + i5
# V = R2i2 + R5i5
# i2 + i3 = i5
# R4i4 = R3i3 + R5i5

A = np.array([[R1,-R2,R3,0,0],[1,1,0,-1,-1],[0,R2,0,0,R5],[0,1,1,0,-1],[0,0,R3,-R4,R5]])
b = np.array([0,0,V,0,0])
print(A)
print(b)
# check the eqns are all independent
det = np.linalg.det(A)
print(det)

x = MyGauss(A,b)
print(x)

# I = i1 + i2
I = x[0]+x[1]
# V = R * I
R = V/I
print(R)
```
